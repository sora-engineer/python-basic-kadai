```python
class Human:
  def __init__(self, name, age):
    self.name = name
    self.age = age

  def check_adult(self):
    if self.age >= 20:
      print(f"{self.name}さんは{self.age}歳であり、大人です。")
    else:
      print(f"{self.name}さんは{self.age}歳であり、大人ではありません。")

humans_data = [
    {"name": "侍太郎", "age": 36},
    {"name": "侍二郎", "age": 15},
    {"name": "侍三郎", "age": 20},
    {"name": "侍四郎", "age": 19},
    {"name": "侍五郎", "age": 25}
]

humans = []

for data in humans_data:
    human = Human(data["name"], data["age"])
    humans.append(human)

for human in humans:
    human.check_adult()
```
