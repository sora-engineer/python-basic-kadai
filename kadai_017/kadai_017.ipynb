```python
class Human:
  def __init__(self, name, age):
    self.name = name
    self.age = age

  def check_adult(self):
    if self.age >= 20:
      print(f"{self.name}さんは{self.age}歳であり、大人です。")
    else:
      print(f"{self.name}さんは{self.age}歳であり、大人ではありません。")

# インスタンスをリストに入れる
humans = [
    Human("侍太郎", 36),
    Human("侍二郎", 15),
    Human("侍三郎", 20),
    Human("侍四郎", 19),
    Human("侍五郎", 25)
]


for human in humans:
    human.check_adult()
```
